```
import numpy as np
import matplotlib.pyplot as plt
```

```
from mpl_toolkits import mplot3d
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```
def f(x, y):
  return x ** 2 + y ** 2

def zero(x, y):
  return x * 0.0 + y * 0.0
```

```
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```
fig = plt.figure(figsize=(8,6))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, zero(X, Y), alpha=0.5)
ax.contour3D(X, Y, Z, 50, alpha=0.5)
ax.scatter([0], [0], [0], marker='o', color='red')
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('l(a,b)')
```
